```python
import math

class Rectangle:
    def __init__(self, ancho, alto):
        self.ancho = ancho
        self.alto = alto

    def area(self):
        return self.alto * self.ancho

class Circle:
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * self.radio ** 2

#q = int(input())
q = 2

for _ in range(q):
    parts = input().split()
    shape_type = parts[0]
    params = list(map(int, parts[1:]))

    if shape_type == "rectangle":
        shape = Rectangle(*params)
    elif shape_type == "circle":
        shape = Circle(*params)

    print(f"{shape.area():.2f}")
```

```python
import os

class VendingMachine:
    def __init__(self, num_items, item_price):
        self.num_items = num_items
        self.item_price = item_price

    def buy(self, req_items, money):
        if req_items > self.num_items:
            raise ValueError("Not enough items in the machine")
        cost = req_items * self.item_price
        if money < cost:
            raise ValueError("Not enough coins")
        self.num_items -= req_items
        return money - cost
    
# Simulación del Sample Input 0

machine = VendingMachine(10, 2)  # 10 items, each costs 2
operations = [
    (1, 5),      # OK → devuelve 3
    (10, 100),   # Error → Not enough items
    (7, 100),    # OK → devuelve 86
    (2, 3),      # Error → Not enough coins
]

for req_items, money in operations:
    try:
        result = machine.buy(req_items, money)
        print(result)
    except ValueError as e:
        print(e)
```
